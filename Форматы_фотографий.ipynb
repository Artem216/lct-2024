# Форматы предоставленных изображений

### Чтобы выбрать форматы, которые мы будем предлагать как возможные для генерации, нужно было рассмотреть какие форматы встречаются в предоставленных данных и с какой частотой.

```python
import cv2
import os
```

```python
shapes = []
for i in os.listdir('/content/drive/MyDrive/ЛЦТ 2024 Генерация/баннеры для хакатона'):
  img = cv2.imread('/content/drive/MyDrive/ЛЦТ 2024 Генерация/баннеры для хакатона/' + i)
  shapes += [img.shape]
print(set(shapes))
```

```python
import matplotlib.pyplot as plt

def show_samples(size):
  imgs = []
  names = []
  for i in os.listdir('/content/drive/MyDrive/ЛЦТ 2024 Генерация/баннеры для хакатона'):
    img = cv2.imread('/content/drive/MyDrive/ЛЦТ 2024 Генерация/баннеры для хакатона/' + i)
    if img.shape == size:
      imgs += [img]
      names += [i]

  fig, ax = plt.subplots(2, 2)
  fig.suptitle(f'Размер: {size}, Количество: {len(imgs)}, Название: {names[-1]}')
  ax[0, 0].imshow(imgs[0])
  ax[0, 1].imshow(imgs[1])
  ax[1, 0].imshow(imgs[2])
  ax[1, 1].imshow(imgs[3])
```

```python
show_samples((320, 580, 3))
```
